### Co2 emission per country from 1750 - 2020 
#### Source: https://www.icos-cp.eu/science-and-impact/global-carbon-budget/2021

```python
from cachetools import TTLCache
from bs4 import BeautifulSoup
import pandas as pd 
import requests

from itertools import chain
import re
import time
```

```python
df = pd.read_csv("co-emissions-per-capita.csv")
df['year'] = df['Year'].astype(str)
```

```python
df['idx'] = df.groupby('year').cumcount()
```

```python
#pivot table 
df = df.pivot(index='Entity',columns='year')[['Annual CO2 emissions (per capita)']]
```

```python
df.to_csv('co24.csv', index=True)
```
